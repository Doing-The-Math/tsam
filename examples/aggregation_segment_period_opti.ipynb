Import pandas and the relevant time series aggregation class

```python
%load_ext autoreload
%autoreload 2
import copy
import os
import pandas as pd
import matplotlib.pyplot as plt
import tsam.timeseriesaggregation as tsam
import tsam.hyperparametertuning as tune
import tqdm
%matplotlib inline
```

### Input data 

Read in time series from testdata.csv with pandas

```python
raw = pd.read_csv('testdata.csv', index_col = 0)
raw=raw.rename(columns={'T': 'Temperature [°C]', 'Load':'Load [kW]', 'Wind':'Wind [m/s]', 'GHI': 'Solar [W/m²]'})
```

Create a plot function for a visual comparison of the time series

```python
def plotTS(plot_data, raw_data, periodlength=24):
    fig, axes = plt.subplots(figsize = [7, 6], dpi = 100, nrows = raw_data.shape[1], ncols = 1)
    for i, column in enumerate(raw.columns):
        data = plot_data[column]
        stacked, timeindex = tsam.unstackToPeriods(copy.deepcopy(data), periodlength)
        cax = axes[i].imshow(stacked.values.T, interpolation = 'nearest', vmin = raw_data[column].min(), vmax = raw_data[column].max(), origin='lower')
        axes[i].set_aspect('auto')  
        axes[i].set_ylabel('Hour')
        plt.xlabel('Day in the year')

        cbar=plt.colorbar(cax, ax=axes[i], pad=0.01, aspect=7)
        cbar.set_label(column)

    fig.subplots_adjust(right = 1.1, hspace = 0.05)
```

Plot an example series - in this case the temperature

```python
plotTS(raw,raw,periodlength=24)
```

### Tune a hierarchical aggregation with segments in combination with duration representation

```python
tunedAggregations = tune.HyperTunedAggregations(
    tsam.TimeSeriesAggregation(
        raw,
        hoursPerPeriod=24,
        clusterMethod="hierarchical",
        representationMethod="durationRepresentation",
        distributionPeriodWise=False,
        rescaleClusterPeriods=False,
        segmentation=True,
    )
)
```

## Now show it for a parameter variation as animation

```python
from matplotlib.animation import FuncAnimation
import numpy as np
```

```python
fig, axes = plt.subplots(figsize = [7, 6], dpi = 100, nrows = raw.shape[1], ncols = 1)
for i, column in enumerate(raw.columns):
    data = raw[column]
    stacked, timeindex = tsam.unstackToPeriods(copy.deepcopy(data), tunedAggregations.base_aggregation.hoursPerPeriod)
    cax = axes[i].imshow(stacked.values.T, interpolation = 'nearest', vmin = raw[column].min(), vmax = raw[column].max(), origin='lower')
    axes[i].set_aspect('auto')  
    axes[i].set_ylabel('Hour')
    plt.xlabel('Day in the year')
    

    cbar=plt.colorbar(cax, ax=axes[i], pad=0.01, aspect=7)
    cbar.set_label(column)

fig.suptitle('Original',y=0.93)
fig.subplots_adjust(right = 1.1, hspace = 0.05)

# chosen set of aggregations TODO make it more systematically
iterator=np.logspace(1, len(tunedAggregations.aggregationHistory), num=40, base=1.0154859)[::-1]-1

def animate(i):
    i=int(iterator[i])
    predictedPeriods = tunedAggregations.aggregationHistory[i].predictOriginalData()
    reduction = 1 -  ( float(tunedAggregations._segmentHistory[i] * tunedAggregations._periodHistory[i]) /len(raw))
    fig.suptitle('Time steps reduction of ' + str(round(reduction*100,2)) + "%" , # " and RMSE of " + str(round(tunedAggregations.aggregationHistory[i].totalAccuracyIndicators()["RMSE"]*100,3)) + "%" 
        y=0.93)
    for i, column in enumerate(raw.columns):
        data = predictedPeriods[column]
        stacked, timeindex = tsam.unstackToPeriods(copy.deepcopy(data), tunedAggregations.base_aggregation.hoursPerPeriod)
        cax = axes[i].imshow(stacked.values.T, interpolation = 'nearest', vmin = raw[column].min(), vmax = raw[column].max(), origin='lower')
        axes[i].set_aspect('auto')  
    fig.subplots_adjust(right = 1.0, hspace = 0.05)


ani = FuncAnimation(fig, animate, frames=len(iterator))

from IPython.display import HTML
HTML(ani.to_jshtml())
```
